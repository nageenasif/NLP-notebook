Sometimes, you will need same operations performed on multiple tasks. It is a good programming exercise to form them into Python functions with meaningful names for further usage. Now you only have to type a short name instead of one or more complete lines of Python code, and you can re-use it as often as you like. The block of code that does a task for us is called a function, and we define a short name for our function with the keyword def. The next example shows how to define two new functions.

<h3> Collocations and Bigrams 

A collocation is a series of exceptionally frequently occurring phrases together. Red wine is therefore a collocation, but the wine is not. A feature of collocations is that they are resistant to replacing words with comparable senses; for instance, maroon wine certainly sounds strange. We begin by extracting a list of word pairs from a text, also known as bigrams.

<h3> Counting Other Things

It is helpful to count words, but we can count other stuff as well. For example, by creating a FreqDist from a long list of numbers where each number is the length of the corresponding word in the text, we can look at the distribution of word lengths in a text:

````
[len(w) for w in text1]
fdist = FreqDist(len(w) for w in text1)
print(fdist)  
fdist
````

One might wonder how often the distinct word lengths are (e.g., how many length four words appear in the text, there are more length five words than length four words, etc.). This we can do as follows:

````
print('Most common: ', fdist.most_common())
print('Max: ', fdist.max())
print('Frequency of three-letter-word: ', fdist[3])
print('Ratio: ', fdist.freq(3))
````

<h3> Conditionals</h3>

Python supports a wide range of operators for testing the relationship between values.

````
# We can use the operators to select different words from a sentence. 
# Examples for relational operators.  They all use sent7, the first sentence 
# from text7 (Wall Street Journal). 

print('All of the sentence:\n', sent7)

print('\nWords in the sentence with lengths are less than 4:\n', 
      [w for w in sent7 if len(w) < 4])

print('\nWords in the sentence with lengths are less than or equal to 4:\n', 
      [w for w in sent7 if len(w) <= 4])

print('\nWords in the sentence with lengths are equal to 4:\n', 
      [w for w in sent7 if len(w) == 4])

print('\nWords in the sentence with lengths are not equal to 4:\n', 
      [w for w in sent7 if len(w) != 4])
````
